```python
# Replace male with 1 and female with 0
df['label'].replace('female',0,inplace=True)
df['label'].replace('male',1,inplace=True)
```

```python
df[df['label']==1].describe()
```

```python
X = df.iloc[:,0:20]
y = df.iloc[:,[20]]
```

```python
# Splitting the data set in test set and training set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#Applying Standard Scaling to the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting the random forest classification to the training set
from sklearn.ensemble import RandomForestClassifier
cs = RandomForestClassifier(n_estimators=200,criterion='entropy',random_state=23)
cs.fit(X_train,y_train)
```

```python
# Predicting the Test set results
y_pred = cs.predict(X_test)
```

```python
# Generating the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, y_test)
```

```python
# Visualizing the confusion matrix
fig, ax = plt.subplots(figsize = (5,5))
sns.heatmap(cm.T, annot=True, annot_kws={"size": 15}, cmap="Oranges", vmin=0, vmax=800, fmt='.0f', linewidths=1, linecolor="white", cbar=False,
           xticklabels=["Female","Male"], yticklabels=["Female","Male"])
plt.ylabel("Predicted", fontsize=15)
plt.xlabel("Actual", fontsize=15)
ax.set_xticklabels(["Female","Male"], fontsize=13)
ax.set_yticklabels(["Female","Male"], fontsize=13)
plt.title("Confusion Matrix for Random forest classifier", fontsize=15)

plt.show()
```

```python
#generating a report to extract the measure of interest using built-in sklearn function
report = classification_report(y_test,y_pred,output_dict = True)
print("\nClassification report for random forest classifier:\n")
print("\tAccuracy = {0:0.3f}".format(report["accuracy"]))
print("\tPrecision = {0:0.3f}".format(report["1"]["precision"]))
print("\tSpecificity = {0:0.3f}".format(report["0"]["recall"]))
print("\tSensitivity = {0:0.3f}".format(report["1"]["recall"]))
print("\tF1-score = {0:0.3f}\n".format(report["1"]["f1-score"]))
```
